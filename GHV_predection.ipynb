```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Data Collection And Analysis

```python
#loading the data from csv file to a pandas dataframe
GHV_dataset = pd.read_csv('/content/GHI_Report_ .csv')
GHV_dataset
```

```python
#Distribution of age value
sns.catplot(x = 'H_Score', data = GHV_dataset, kind = 'count')
```

```python
# seperating the data and label
X = GHV_dataset.drop(columns = 'H_Score', axis = 1)
```

```python
Y = GHV_dataset.H_Score
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,test_size =0.2 , random_state =2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training
```

```python
# Calling Linear Regression Model
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
# predection on training the Data
training_data_predection = regressor.predict(X_train)
```

```python
# R Square Value
r2_train = metrics.r2_score(Y_train,training_data_predection)
print( "R Square Value " , r2_train )
```

```python
# R square value on testing data

testing_data_prediction = regressor.predict(X_test)

r2_test = metrics.r2_score(Y_test,testing_data_prediction)
print( "R Square Value " , r2_test )
```
